```python
import pandas as pd
import matplotlib.pyplot as plt

#  
file_url = "https://drive.google.com/uc?id=1XjQfYUyoqUf-0sa0fZ082FUuMcuar2aG"

#     
spotify_data = pd.read_csv(file_url, encoding='ISO-8859-1')

#      
spotify_data.head()

#     'streams'    
spotify_data['streams'] = spotify_data['streams'].str.extract(r'(\d+)').astype(float)

#         10 
top_songs = spotify_data.nlargest(10, 'streams')

#    
plt.figure(figsize=(10, 6))
plt.barh(top_songs['track_name'], top_songs['streams'], color='skyblue')
plt.xlabel(' ')
plt.ylabel(' ')
plt.title('10    Spotify  2023 ')
plt.gca().invert_yaxis()
plt.show()

#     danceability
plt.figure(figsize=(8, 8))
plt.pie(spotify_data['danceability_%'], labels=spotify_data.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('  danceability     Spotify  2023 ')
plt.show()

```

```python
#        2023
specific_track_data = spotify_data[(spotify_data['track_name'] == '  ') & (spotify_data['released_year'] == 2023)]

#   
plt.figure(figsize=(10, 6))
plt.plot(specific_track_data['released_month'], specific_track_data['streams'], marker='o', linestyle='-')
plt.xlabel('')
plt.ylabel(' ')
plt.title('       2023')
plt.xticks(range(1, 13))  #       X
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#     2023
spotify_data_2023 = spotify_data[spotify_data['released_year'] == 2023]

#         10
top_artists = spotify_data_2023.groupby('artist(s)_name')['streams'].sum().nlargest(10)

#   
plt.figure(figsize=(10, 6))
top_artists.plot(kind='bar', color='skyblue')
plt.xlabel('')
plt.ylabel(' ')
plt.title('10    Spotify  2023 ')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#     2023   
specific_tracks_data = spotify_data[(spotify_data['released_year'] == 2023) & spotify_data['track_name'].isin(['Super Shy', 'LALA', 'Seven (feat. Latto) (Explicit Ver.)'])]

#    
plt.figure(figsize=(10, 6))
plt.barh(specific_tracks_data['track_name'], specific_tracks_data['streams'], color='skyblue')
plt.xlabel(' ')
plt.ylabel('')
plt.title('       2023 ')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```
