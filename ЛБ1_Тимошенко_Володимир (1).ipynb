```python
import pandas as pd
import matplotlib.pyplot as plt

# Завантаження файлу
file_url = "https://drive.google.com/uc?id=1XjQfYUyoqUf-0sa0fZ082FUuMcuar2aG"

# Завантаження файлу з іншим кодуванням
spotify_data = pd.read_csv(file_url, encoding='ISO-8859-1')

# Виведення перших рядків даних для перевірки
spotify_data.head()

# Очистка значень в стовпці 'streams' і вибір числової інформації
spotify_data['streams'] = spotify_data['streams'].str.extract(r'(\d+)').astype(float)

# Сортування даних за кількістю прослуховувань та вибір перших 10 записів
top_songs = spotify_data.nlargest(10, 'streams')

# Побудова горизонтальної стовпчикової діаграми
plt.figure(figsize=(10, 6))
plt.barh(top_songs['track_name'], top_songs['streams'], color='skyblue')
plt.xlabel('Кількість прослуховувань')
plt.ylabel('Назва пісні')
plt.title('10 найпопулярніших пісень на Spotify у 2023 році')
plt.gca().invert_yaxis()
plt.show()

# Побудова кругової діаграми за danceability
plt.figure(figsize=(8, 8))
plt.pie(spotify_data['danceability_%'], labels=spotify_data.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Розподіл за danceability серед найпопулярніших пісень на Spotify у 2023 році')
plt.show()

```

```python
# Вибір даних для певного треку та року 2023
specific_track_data = spotify_data[(spotify_data['track_name'] == 'Назва певного треку') & (spotify_data['released_year'] == 2023)]

# Побудова лінійного графіку
plt.figure(figsize=(10, 6))
plt.plot(specific_track_data['released_month'], specific_track_data['streams'], marker='o', linestyle='-')
plt.xlabel('Місяць')
plt.ylabel('Кількість прослуховувань')
plt.title('Зміни кількості прослуховувань певного треку протягом року 2023')
plt.xticks(range(1, 13))  # Встановлення місяців як міток на вісі X
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Вибір даних для року 2023
spotify_data_2023 = spotify_data[spotify_data['released_year'] == 2023]

# Сортування артистів за кількістю слухачів та вибір перших 10
top_artists = spotify_data_2023.groupby('artist(s)_name')['streams'].sum().nlargest(10)

# Побудова стовпчикової діаграми
plt.figure(figsize=(10, 6))
top_artists.plot(kind='bar', color='skyblue')
plt.xlabel('Артист')
plt.ylabel('Кількість прослуховувачів')
plt.title('10 найпопулярніших артистів на Spotify у 2023 році')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Вибір даних для року 2023 та конкретних треків
specific_tracks_data = spotify_data[(spotify_data['released_year'] == 2023) & spotify_data['track_name'].isin(['Super Shy', 'LALA', 'Seven (feat. Latto) (Explicit Ver.)'])]

# Побудова горизонтальної стовпчикової діаграми
plt.figure(figsize=(10, 6))
plt.barh(specific_tracks_data['track_name'], specific_tracks_data['streams'], color='skyblue')
plt.xlabel('Кількість прослуховувань')
plt.ylabel('Трек')
plt.title('Порівняння кількості прослуховувань трьох конкретних треків у 2023 році')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```
